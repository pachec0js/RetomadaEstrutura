# Manipulação de Tipos de Dados com Pandas

Olá, estudante! Nesta aula, vamos aprender na prática a diferença entre dados estruturados (como planilhas) e dados semiestruturados (como arquivos JSON). Também veremos como criar e ler esses tipos de arquivos utilizando a biblioteca `pandas`.

## 1. Dados Estruturados

Dados estruturados são altamente organizados e formatados de maneira que sejam facilmente pesquisáveis em bancos de dados relacionais. Um exemplo clássico são as planilhas do Excel, onde os dados são divididos em linhas e colunas.

No código abaixo, vamos:
1. Criar um **dicionário** em Python contendo informações de pessoas (Nome, Idade e Cidade).
2. Transformar esse dicionário em um `DataFrame` do Pandas (que é uma tabela em formato estruturado).
3. Exportar essa tabela para um arquivo do Excel chamado `dadosEstruturados.xlsx`.

> **Dica:** Para rodar este código, você precisa ter as bibliotecas `pandas` e `openpyxl` instaladas (`pip install pandas openpyxl`).

```python
# Importando a biblioteca pandas
import pandas as pd

# Estrutura de Dicionário
dados_planilha = {
    'Nome': [
        'Fábio Pacheco',
        'Bruno Pezzolato',
        'Arthur Buscarino',
        'Felipe Polegato'
    ],
    'Idade': [
        45,
        34,
        42,
        38
    ],
    'Cidade': [
        'São Paulo',
        'São Caetano do Sul',
        'São Bernardo do Campo',
        'Santo André'
    ]
}

# Criando o DataFrame a partir do dicionário
df_planilha = pd.DataFrame(dados_planilha)

print("Visualizando o DataFrame criado:")
print(df_planilha)

# Salvando o DataFrame em um arquivo Excel
with pd.ExcelWriter('dadosEstruturados.xlsx') as writer:
    df_planilha.to_excel(writer, sheet_name='Planilha Estruturada', index=False)

print('\nPlanilha criada com sucesso!')
```

## 2. Dados Semiestruturados

Os dados semiestruturados não obedecem a uma estrutura formal de modelos de dados associados a bancos de dados relacionais, mas contêm tags ou outros marcadores para separar elementos semânticos. Um excelente exemplo é o formato **JSON** (JavaScript Object Notation), muito utilizado na web.

No código a seguir, usaremos a mesma base de dados, mas desta vez iremos salvá-la no formato `.json`.

```python
import pandas as pd

# Definindo os mesmos dados
dados_json = {
    'Nome': [
        'Fábio Pacheco',
        'Bruno Pezzolato',
        'Arthur Buscarino',
        'Felipe Polegato'
    ],
    'Idade': [
        45,
        34,
        42,
        38
    ],
    'Cidade': [
        'São Paulo',
        'São Caetano do Sul',
        'São Bernardo do Campo',
        'Santo André'
    ]
}

# Criando o DataFrame
df_json = pd.DataFrame(dados_json)

# Exportando para o formato JSON
df_json.to_json('dadosSemi.json', orient='records', lines=False)

print('JSON criado com sucesso!')
```

## 3. Leitura dos Arquivos Criados

Agora que já criamos um arquivo estruturado (`.xlsx`) e um arquivo semiestruturado (`.json`), vamos aprender como ler esses arquivos e trazê-los de volta para o ambiente do Python na forma de DataFrames do Pandas.

Isso é extremamente comum no dia a dia do Cientista de Dados: importar arquivos de fontes e formatos diferentes para realizar a análise!

```python
import pandas as pd

# Lendo o arquivo Excel estruturado
df_excel = pd.read_excel('dadosEstruturados.xlsx', sheet_name='Planilha Estruturada')
print('Planilha Excel importada com sucesso:')
print(df_excel)

# Lendo o arquivo JSON semiestruturado
df_json_lido = pd.read_json('dadosSemi.json')
print('\nArquivo JSON importado com sucesso:')
print(df_json_lido)
```
