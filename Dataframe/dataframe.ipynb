```python
import pandas as pd

dados = {
    'Nome': ['Fábio', 'Maria', 'Carlos', 'Ana', 'Lucas'],
    'Idade': [18, 22, 30, 25, 19],
    'Naturalidade': ['SP', 'AM', 'RJ', 'SP', 'RJ']
}

df=pd.DataFrame(dados)
print(df)
```

```python
# Verificar Coluna Especifica
print(df['Naturalidade'])
```

```python
# Puxar valor com condicional
print(df["Idade"]>20)
```

```python
# Puxar Valor Condicional Com Infos Completas
# 1.Filtrar a idade - f["Idade"]>20
# 2.Puxar todas as infos - df[]
print(df[
    df['Idade']>20
])
```

```python
# Puxar Os Registros Do Fábio
# 1.Filtro print(df['Nome']=='Fábio')

print(df[
    df['Nome']=='Fábio'
])

```

```python
# Agrupar / Calculo de Média de Idade por Naturalidade

media_nat=df.groupby('Naturalidade')['Idade'].mean()

print(media_nat)
```
