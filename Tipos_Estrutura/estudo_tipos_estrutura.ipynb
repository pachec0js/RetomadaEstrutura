# Estruturas de Dados em Python

Olá, estudante! Neste notebook, vamos explorar na prática as principais estruturas de dados nativas do Python: **Listas, Tuplas, Conjuntos (Sets) e como implementar uma Fila básica**. 

Essas estruturas são fundamentais para armazenar e organizar dados de forma eficiente nos seus programas.

## 1. Listas

A **Lista** é uma estrutura de dados mutável, ou seja, você pode alterar, adicionar e remover elementos após sua criação. Ela é ordenada e permite elementos duplicados.

Abaixo, vemos como criar uma lista e utilizar os métodos mais comuns como `append`, `pop`, `insert`, `remove`, `sort` e `reverse`.

```python
# Criando uma lista de frutas
frutas = ["banana", "maca", "uva"]

# Adicionando um elemento ao final da lista
frutas.append("cereja")
print("Lista atualizada após append:", frutas)

# Removendo o último elemento da lista
removido = frutas.pop()
print("Removido:", removido)
print("Lista atualizada após pop:", frutas)

# Inserindo um elemento em uma posição específica (índice 0)
frutas.insert(0, "laranja")
print("Elemento na posição 0:", frutas[0])
print("Lista atualizada após insert:", frutas)

# Removendo um elemento específico pelo seu valor
frutas.remove("maca")
print("Lista atualizada após remove:", frutas)

# Ordenando a lista em ordem alfabética
frutas.sort()
print("Lista ordenada:", frutas)

# Invertendo a ordem dos elementos da lista
frutas.reverse()
print("Lista invertida:", frutas)
```

## 2. Tuplas

A **Tupla** é semelhante à lista, com uma diferença crucial: ela é **imutável**. Depois de criar uma tupla, você não pode adicionar, remover ou alterar seus elementos. 

Elas são excelentes para dados que não devem ser modificados durante a execução do programa.

```python
# Criando uma tupla de cores (usamos parênteses em vez de colchetes)
cores = ('vermelho', 'azul', 'verde')

print(f"Tupla inteira: {cores}")

# Acessando elementos pelo índice
print(f"Primeira cor (índice 0): {cores[0]}")
print(f"Última cor (índice -1): {cores[-1]}")

# Obtendo o tamanho da tupla
print(f"Quantidade de cores na tupla: {len(cores)}")
```

## 3. Conjuntos (Sets)

Os **Conjuntos** são coleções não ordenadas e **não permitem elementos duplicados**. São muito úteis para operações matemáticas de conjuntos, como união, interseção e diferença.

```python
# Criando conjuntos (usamos chaves, mas sem estrutura de chave-valor como dicionários)
conjunto_a = {1, 2, 3, 4}
conjunto_b = {3, 4, 5, 6, "Olá"}

print(f"Conjunto A: {conjunto_a}")
print(f"Conjunto B: {conjunto_b}\n")

# União (juntar os dois conjuntos sem repetir elementos)
uniao = conjunto_a.union(conjunto_b)
uniao2 = conjunto_a | conjunto_b
print(f"União (método union): {uniao}")
print(f"União (operador |): {uniao2}\n")

# Interseção (apenas os elementos comuns aos dois conjuntos)
intersecao = conjunto_a.intersection(conjunto_b)
intersecao2 = conjunto_a & conjunto_b
print(f"Interseção (método intersection): {intersecao}")
print(f"Interseção (operador &): {intersecao2}\n")

# Diferença (elementos que estão no A mas NÃO estão no B)
diferenca = conjunto_a.difference(conjunto_b)
diferenca2 = conjunto_a - conjunto_b
print(f"Diferença A-B (método difference): {diferenca}")
print(f"Diferença A-B (operador -): {diferenca2}\n")

# Diferença Simétrica (elementos que estão em A ou B, mas não em ambos)
simetrica = conjunto_a.symmetric_difference(conjunto_b)
simetrica2 = conjunto_a ^ conjunto_b
print(f"Diferença Simétrica (método symmetric_difference): {simetrica}")
print(f"Diferença Simétrica (operador ^): {simetrica2}")
```

## 4. Filas (Queues)

Uma **Fila** é uma estrutura que segue o conceito **FIFO (First In, First Out)** - o primeiro a entrar é o primeiro a sair (como uma fila de banco).

Em Python, podemos simular uma fila de forma simples usando uma lista convencional e controlando a entrada (no final) e a saída (no início).

```python
# Criando uma fila vazia (usando lista)
fila = []

# Para INSERIR na fila, usamos o append() - insere no final
fila.append(10)
fila.append(20)
fila.append(30)
fila.append(40)

print('Fila após inserções:', fila)

# Para REMOVER da fila, usamos o pop(0) - remove o primeiro elemento (índice 0)
retirar = fila.pop(0)

print('Elemento retirado da fila:', retirar)
print('Fila atualizada:', fila)
```
